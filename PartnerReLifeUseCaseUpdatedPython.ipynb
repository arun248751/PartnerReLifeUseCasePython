Problem 1  Data handling, analysis and plotting
The first problem of the case study builds on the data in the files p01-02_portfolio.csv and
p01-02_rates.csv. One file contains membership information for a Group Life portfolio and one
has information on the rates which should be charged.

Question a.

1.   Read the data from the two files into Rs memory. The rates are applicable to each individual in the portfolio,
depending on that individuals age and gender.
2.   Combine the two datasets into a single table by looking up
the rate for each line of the portfolio.




````
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb

from sklearn.metrics import (
    roc_auc_score, roc_curve, precision_recall_curve, average_precision_score,
    confusion_matrix, classification_report, brier_score_loss
)
from sklearn.calibration import calibration_curve

# For SMOTE resampling
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline

# For SHAP explainability (optional, may be slow on large data)
import shap
````

````
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# === Load Data ===
portfolio = pd.read_csv("/content/p01-02_portfolio.csv",sep=";")
rates = pd.read_csv("/content/p01-02_rates.csv",sep=";")
````

````
portfolio.columns
````

````
portfolio.head()
````

````
# === Light EDA ===
print("Portfolio head:\n", portfolio.head())
print("\nRates head:\n", rates.head())
print("\nPortfolio Missing Values:\n", portfolio.isna().sum())
````

````
# Convert date to datetime
portfolio["Date.of.Birth"] = pd.to_datetime(portfolio["Date.of.Birth"], format="%d.%m.%Y", errors="coerce")
````

````
# Calculate AGE (simple version: today - birthyear)
portfolio["Age"] = 2025 - portfolio["Date.of.Birth"].dt.year
````

````
print("Portfolio head:\n", portfolio.head())
print("\nRates head:\n", rates.head())
print("\nPortfolio Missing Values:\n", portfolio.isna().sum())
````

````
# Merge portfolio + rates by age & gender
combined = portfolio.merge(
    rates,
    how="left",
    left_on=["Age", "Gender"],
    right_on=["Age", "Gender"]
)
````

````
# Clean Death Sum Insured (DeathSI)
combined["DeathSI"] = pd.to_numeric(combined["DeathSI"], errors="coerce")
````

````
combined.head()
````

````
# Replace missing DeathSI with median of same scheme
combined["DeathSI"] = combined.groupby("SchemeName")["DeathSI"].transform(
    lambda x: x.fillna(x.median())
)

# Fill any remaining DeathSI NaNs (from schemes where all were NaN) with the overall median
combined["DeathSI"].fillna(combined["DeathSI"].median(), inplace=True)

# Fill NaN rates with 0, assuming no claims for unrated ages
combined["Rate"].fillna(0, inplace=True)

print(combined.head())
````

````
# -----------------------------------
# Problem 1C  Monte Carlo simulation
# -----------------------------------

import seaborn as sns

np.random.seed(1234)
nsim = 1000

results = []

schemes = combined["SchemeName"].unique()

for scheme in schemes:
    subset = combined[combined["SchemeName"] == scheme]
    n = subset.shape[0]

    # Run 1000 simulations
    cost_list = []
    for _ in range(nsim):
        # Probability of claim = Rate / 1000
        claim_flag = np.random.rand(n) < (subset["Rate"] / 1000)
        x = claim_flag * subset["DeathSI"]
        cost_list.append(sum(x))

    results.append(pd.DataFrame({
        "Scheme": scheme,
        "Cost": cost_list
    }))

sim_results = pd.concat(results)
````

**Question b. The variable `group` has different factor levels in the data and scoring set. How will this difference impact your predictive model? What solutions might you offer?**

**Impact on your Predictive Model:**
If the `group` variable in the scoring (test) set contains categories (factor levels) that were not present in the training data, or vice-versa, it can significantly impact the predictive model:
1.  **Preprocessing Failure:** Without proper handling, the One-Hot Encoder, which transforms categorical features into numerical ones, would encounter an unseen category. This could lead to an error during the `transform` step, as it wouldn't know how to create a binary column for a new category.
2.  **Inconsistent Feature Space:** Even if an error doesn't occur, the resulting feature set (number and order of columns) generated by the encoder might differ between the training and scoring data. This inconsistency would prevent the trained model from making predictions, as it expects the same features it was trained on.
3.  **Poor Predictions:** If new categories are encountered and handled (e.g., by creating all-zero vectors for them), the model will not have learned any specific patterns or relationships for these categories. **Solutions Offered:**
1.  **Handle Unknown Categories During Encoding (`OneHotEncoder(handle_unknown='ignore')`):** This is the most straightforward and common solution. When using `OneHotEncoder`, setting `handle_unknown='ignore'` instructs the encoder to create all-zero vectors for any category encountered during transformation that was not seen during the `fit` phase (i.e., in the training data). This ensures a consistent number of output features.
2.  **Align Categorical Levels Explicitly:** Before applying the encoder, explicitly define all possible categories for the `group` variable by combining the unique levels from both the training and scoring datasets. Then, convert the `group` columns in both `train` and `test` to `Categorical` dtype using these `combined_levels`. This ensures that the encoder sees all potential categories during `fit` and `transform`, preventing 'unknown' categories in the first place (as was done in cell `x8HyUZXbCic_`).
3.  **Group Sparse/Rare Categories:** If some categories are very rare (appearing only a few times or only in one of the datasets), they could be grouped into an 'Other' category before encoding. This reduces the dimensionality and prevents the model from overfitting to noisy or sparse categories.

### Interpretation of Calibration Plots

Calibration plots (also known as reliability diagrams) visualize how well the predicted probabilities of a classification model align with the actual probabilities of the positive class. They are crucial for understanding the trustworthiness of the model's probabilistic outputs.

#### **Statistical Interpretation**

1.  **What it is:** A calibration plot shows the fraction of positives (actual probability) for different predicted probability bins. Ideally, if a model is perfectly calibrated, its predictions should fall along the diagonal line (y=x), meaning a predicted probability of 0.70 corresponds to the positive class occurring 70% of the time in that bin.
2.  **How to interpret:**
    *   **Diagonal Line (y=x):** Represents perfect calibration.
    *   **Above the diagonal:** The model is under-predicting the probability of the positive class. For instance, if the model predicts a 40% chance, but the positive class actually occurs 60% of the time, the model is under-confident.
    *   **Below the diagonal:** The model is over-predicting the probability of the positive class. If the model predicts an 80% chance, but the positive class only occurs 60% of the time, the model is over-confident.
3.  **Brier Score:** The Brier score (lower is better) is a single metric that quantifies the overall calibration error, integrating both accuracy and calibration. It represents the mean squared difference between the predicted probability and the actual outcome (0 or 1).

#### **Business Interpretation (Reinsurance Context)**

1.  **Reliability of Probabilities for Decision Making:**
    *   **Crucial for Quantitative Decisions:** In reinsurance, many decisions are based on probabilities: setting reserves, pricing treaties, and quantifying risk. If a model predicts a claim probability of 10%, the business needs to trust that this means, out of all similar cases, about 10% will actually result in a claim. Poorly calibrated probabilities can lead to severe miscalculations in financial planning.
    *   **Risk Quantification:** For instance, if `outcome=1` is a large claim event, a well-calibrated model allows the reinsurer to accurately quantify the expected loss associated with a portfolio or individual policy. An uncalibrated model might systematically understate or overstate risk, leading to inadequate capital reserves or uncompetitive pricing.

2.  **Trust and Acceptance:**
    *   **Stakeholder Confidence:** Business stakeholders (underwriters, actuaries, risk managers) are more likely to trust and adopt a model if its probabilities are reliable. A visual calibration plot provides a clear, intuitive way to demonstrate this reliability.

3.  **Financial Impact of Miscalibration:**
    *   **Under-prediction (Above diagonal):** If the model consistently under-predicts the probability of claims, the reinsurer might set insufficient premiums or hold inadequate reserves, exposing them to unexpected losses.
    *   **Over-prediction (Below diagonal):** If the model consistently over-predicts, the reinsurer might charge excessively high premiums (making them uncompetitive) or hold unnecessary capital, impacting efficiency.

4.  **Comparison Across Models:**
    *   **Model Selection:** Comparing the calibration curves and Brier scores (as shown in cell `3aTZDLSqJpuu`) helps select the model that provides the most trustworthy probabilities. In this case, XGBoost, with its lowest Brier score, likely provides the most reliable probability estimates, which is a significant advantage for any business making quantitative risk decisions.

### Interpretation of Thresholding and Classification Metrics

Cell `PdaqD_QUJFoF` provides a detailed look at the classification performance of each model (Baseline Logistic, Random Forest, and XGBoost) at two different thresholds: the default 0.5 threshold and the Youden's J threshold. These metrics are crucial for understanding the practical implications of your models.

#### **Statistical Interpretation**

1.  **Youden's Threshold:**
    *   **What it is:** Youden's J statistic (Sensitivity + Specificity - 1) is a common metric used to select an optimal probability cutoff point for a diagnostic test. The threshold that maximizes Youden's J statistic is chosen as it optimizes the balance between True Positive Rate (Sensitivity) and True Negative Rate (Specificity).
    *   **Interpretation:** This threshold aims to maximize the overall correctness of the model's classifications, giving equal weight to correctly identifying both positive and negative cases. It might differ significantly from the default 0.5, especially in imbalanced datasets.

2.  **Confusion Matrix:**
    *   **What it is:** A confusion matrix is a table used to describe the performance of a classification model on a set of test data for which the true values are known. It has four key components:
        *   **True Positives (TP):** Cases correctly identified as positive.
        *   **True Negatives (TN):** Cases correctly identified as negative.
        *   **False Positives (FP):** Cases incorrectly identified as positive (Type I error).
        *   **False Negatives (FN):** Cases incorrectly identified as negative (Type II error).
    *   **Interpretation:** By examining the counts in each cell, we can understand the types of errors each model makes. For instance, a high number of FN might be acceptable if FP are very costly, or vice-versa.

3.  **Metrics (@0.5 and @Youden Threshold):**
    *   **Precision:** The proportion of positive identifications that were actually correct (TP / (TP + FP)). High precision means fewer false positives.
    *   **Recall (Sensitivity):** The proportion of actual positives that were identified correctly (TP / (TP + FN)). High recall means fewer false negatives.
    *   **F1-Score:** The harmonic mean of precision and recall (2 * (Precision * Recall) / (Precision + Recall)). It provides a single score that balances both metrics, especially useful for imbalanced datasets.
    *   **Accuracy:** The proportion of total predictions that were correct ((TP + TN) / (TP + FP + TN + FN)).
    *   **Interpretation:** These metrics provide a detailed breakdown of model performance. Comparing them across thresholds highlights the trade-offs. For example, moving from 0.5 to Youden's threshold might increase recall for the minority class at the expense of precision or vice-versa, depending on the model and data.

#### **Business Interpretation (Reinsurance Context)**

1.  **Threshold Selection (0.5 vs. Youden):**
    *   **Strategic Decision:** The choice of threshold is a critical business decision, directly impacting operational strategies and financial outcomes. The default 0.5 threshold might be suitable when positive and negative errors have roughly equal costs. However, Youden's threshold offers an alternative that optimizes overall diagnostic performance.
    *   **Reinsurance Implications:** For a reinsurer, if `outcome=1` represents a high-cost claim event:
        *   A threshold leading to high **Recall** (low FN) is desirable if missing a high-cost claim is extremely detrimental (e.g., severe financial impact, reputation damage). This means the reinsurer wants to correctly identify as many potential claims as possible, even if it means reviewing more false positives.
        *   A threshold leading to high **Precision** (low FP) is preferred if investigating or reserving for false high-cost claims is very expensive (e.g., wasted actuarial resources, unnecessary capital allocation). This means the reinsurer wants to be very sure that a flagged claim is indeed going to happen.

2.  **Confusion Matrix & Cost of Errors:**
    *   **Quantifying Financial Impact:** Each cell in the confusion matrix (TP, TN, FP, FN) can be associated with a specific financial or operational cost/benefit for the reinsurer. For example, a False Negative (missing a high-cost claim) might result in significant unexpected payouts, while a False Positive (identifying a low-risk policy as high-risk) might lead to unnecessary scrutiny or over-reserving.
    *   **Operational Efficiency:** The matrix guides operational adjustments. If the model generates too many FP, the reinsurance underwriting team might be overwhelmed. If too many FN, the claims team might be blindsided.

3.  **F1-Score for Imbalanced Data:**
    *   **Focus on Minority Class:** In reinsurance, `outcome=1` (e.g., a catastrophic event, a very large claim) is often the minority class. The F1-score is particularly relevant here because it provides a balanced measure of the model's ability to correctly classify this crucial minority class. A higher F1-score means the model is better at identifying the relevant, often rare, events that have the most significant business impact.
    *   **Overall Model Utility:** Comparing F1-scores across models and thresholds helps determine which configuration offers the best practical utility for the reinsurer's specific objectives, given the typical imbalance of claims data.

````
from sklearn.metrics import f1_score

def youden_threshold(y_true, probs):
    fpr, tpr, thr = roc_curve(y_true, probs)
    J = tpr - fpr
    idx = np.argmax(J)
    return thr[idx]

for name, r in results.items():
    prob = r["proba"]
    # threshold 0.5
    pred_05 = (prob >= 0.5).astype(int)
    # Youden's threshold
    thr_y = youden_threshold(y_hold, prob)
    pred_y = (prob >= thr_y).astype(int)
    print("----", name, "----")
    print("Youden threshold:", thr_y)
    print("Confusion matrix @0.5:\n", confusion_matrix(y_hold, pred_05))
    print("Metrics @0.5:\n", classification_report(y_hold, pred_05, digits=3))
    print("F1 @0.5:", f1_score(y_hold, pred_05))
    print("Confusion matrix @Youden:\n", confusion_matrix(y_hold, pred_y))
    print("Metrics @Youden:\n", classification_report(y_hold, pred_y, digits=3))
    print("F1 @Youden:", f1_score(y_hold, pred_y))
    print()
````
